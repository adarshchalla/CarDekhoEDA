```python
body_type_list=['sedan', 'hatchback', 'suv', 'luxury', 'muv', 'minivan', 'convertible', 'coupe', 'hybrid', 'wagon', 'super-luxury', 'hybrid']
transmission_list= ['manual','automatic']
seller_type_list= ['dealer','individual']
owner_type_list= ['unregistered','first-owner','second-owner','third-owner-or-more']
colors_list= ['white', 'silver', 'red', 'blue', 'black', 'brown', 'orange', 'green', 'violet', 'yellow', 'gold', 'other']
city_list=['bangalore']
```

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import requests
from bs4 import BeautifulSoup

import re
import time

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'


class_for_carname = 'title'
class_for_fuel = 'dotlist truncate'
class_for_price = 'price'

all_carname = []
all_fuel = []
all_car_price = []
bodytype=[]
transmission=[]
seller_type=[]
owner_type=[]
city=[]
color=[]

for l in city_list:

    for j in transmission_list:
        for k in body_type_list:
            for i in colors_list:
                for m in owner_type_list:
                    for n in seller_type_list:
                    
                        url = "https://www.cardekho.com/used-"+str(i)+str('+')+str(j)+str('+')+str(k)+str('+')+str('cars')+str('+')+str('in')+str('+')+str(l)+str('+')+str(m)+str('+')+str(n)
                        agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
                        page = requests.get(url, headers=agent)
                        soupdata = BeautifulSoup(page.text)
                        print(url)
                        for x in soupdata.find_all('div', attrs={'class':'gsc_col-sm-12 gsc_col-md-12 trustmarkbox holder'}):
                            carname = x.find('div',attrs={'class':class_for_carname}).text
                            carprice = x.find('div',attrs={'class':class_for_price}).text
                            carfuel = x.find('div',attrs={'class':class_for_fuel}).text
                            this_carfeatures = ""
                            all_carname.append(carname)
                            all_fuel.append(carfuel.replace(",",""))
                            all_car_price.append((carprice.replace("*","").replace('Rs.','').replace(' Lakh','').replace(",",'').replace(" Cr",'00')))
                            
                            
                            bodytype.append(k)
                            transmission.append(j)
                            seller_type.append(n)
                            owner_type.append(m)
                            city.append(l)
                            color.append(i)
 
df2 = pd.DataFrame({'CarName':all_carname,'city':city, 'Fuel_type':all_fuel, 'car_price':all_car_price, 'body_type':bodytype, 'transmission':transmission, 'seller_type':seller_type, 'owner_type':owner_type, 'car_price':all_car_price, 'color':color})
```

```python
import re

regex = r'(^\d+)'
df2['Year'] = df2['CarName'].apply(lambda x:re.compile(regex).findall(x))


regex = r'([a-zA-Z-]+).*'
df2['Brand'] = df2['CarName'].apply(lambda x:re.compile(regex).findall(x))


regex = r'[0-9,]+\s+\w+\s+(\w+)'
df2['Fuel_used'] = df2['Fuel_type'].apply(lambda x:re.compile(regex).findall(x))

regex = r'\d+'
df2['km_driven'] = df2['Fuel_type'].apply(lambda x:re.compile(regex).findall(x))

df2.to_csv('carwale_bangalore.csv')
```
